```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis', model = 'distilbert-base-uncased-finetuned-sst-2-english')
```

```python
# Test 1: Simple Positive
result = classifier("I Love this project. I am learning so much!")
print(result)

# Test 2: Simple Negative
result_neg = classifier("I am stuck and this is very frustrating/.")
print(result_neg)
```

```python
# Test 3: Sarcasm (Models are bad at this)
# This is a negative statement, but the words are positive.
result_sarcasm = classifier("Wow, this broken app is *just* what I needed today.")
print(f"Sarcasm Test: {result_sarcasm}")

# Test 4: Complex Negation
# This is a positive statement.
result_negation = classifier("This movie was not bad at all.")
print(f"Negation Test: {result_negation}")
```
